```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rc("font", size=12)
plt.figure(figsize=(3.5, 5))
plt.subplot(211)

training_accuracy,validation_accuracy = np.genfromtxt("../data/vgg_like_history.txt").T


plt.plot(training_accuracy, label='training data')
plt.plot(validation_accuracy, label='validation data')
plt.ylabel("accuracy")
plt.legend(frameon=0)
plt.text(0.05, 0.86, "VGG", transform=plt.gca().transAxes)


plt.subplot(212)
training_accuracy,validation_accuracy = np.genfromtxt("../data/resnet_history.txt").T   
plt.plot(training_accuracy, label='training data')
plt.plot(validation_accuracy, label='validation data')
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend(frameon=0)
plt.text(0.05, 0.86, "ResNet", transform=plt.gca().transAxes)
#plt.savefig("accuracy_all.pdf", bbox_inches='tight', transparent=True)
```
